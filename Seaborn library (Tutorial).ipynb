# Scatter Plots

The Scatter plot is a mainstay of statistical visualization. It depicts the joint distribution of two variables using a clodu of points , where each point represents an observation in the dataset. This depiction allows the eye to infer a substantial amount of information about wether there is any meaningful relationship between them.

Although, scatter plot can be plotted for both numerical as well as categorical data, it's mostly used to depict relationship between numerical data.

There are 2 ways to draw a scatterplot in seaborn 
1. replot [ Figure level function]
2. scatterplot [axes level function]

as we will see, these visualization can be quite helpful because they use simply and easily- understood representation of data even on complex data by mapping up to three additional variables using the semantics of hue,size and style.


```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
```

```python
data=sns.load_dataset('tips')
```

```python
data.head()
```

# Plot a scatter plot between total_bill and tip

```python
# plot here
sns.relplot(x='total_bill', y='tip',kind='scatter',data=data)
```

# The hue parameter

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',data=data)
```

# The style parameter

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',data=data)
```

# The size parameter

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',size='size',data=data)
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',size='size',sizes=(15,200),data=data)
```

```python
sns.scatterplot(x='total_bill',y='tip',hue='smoker',data=data)
```

# What is Hue.
<br>
In Seaborn, a Python data visualization library based on Matplotlib, the term "hue" refers to a parameter that allows you to group or categorize data and then apply a different color palette to each group.

## There are also different 'kinds' like:-

Yes, you can use a different kind parameter in the catplot function to create a different type of categorical plot. Here are some other options you can use:

1.kind='strip': This creates a strip plot.

2.kind='swarm': This creates a swarm plot.

3.kind='bar': This creates a bar plot.

4.kind='count': This creates a count plot.

5.kind='point': This creates a point plot.

6.kind='boxen': This creates a boxen plot.

7.kind='violin': This creates a violin plot.

8.kind='boxen': This creates a boxen plot.

9.kind='point': This creates a point plot.

10.kind='line': This creates a line plot.

You can choose the one that best suits your data and the type of visualization you want to create. For example, if you want to create a bar plot instead of a box plot, you would use kind='bar'




# Violin Plots

<b> Violin plots are method of plotting numeric data and can be considered a combination of the box plot with a kernerl density plot.</b>


<b>The advantage of the violin plot over the box plot is that aside from showing the abovementioned statistics it also shows the entire distribution of the data. 

<b>Violin plots are often used to compare the distribution of a given variable across some categories.</b>

```python
sns.violinplot(tips['total_bill'])
```

```python
sns.catplot(x='day',y='total_bill',kind='violin',data=tips)
```

```python
sns.catplot(x='day',y='total_bill',kind='violin',hue='sex',split=True,data=tips)
# split here is used to generate gap because if we won't use it then it will be cluttered i.e. the graphs will not be clearly 
# visible
```

# Heatmap

<b>A heatmap is a graphical representation of data where the individual values contained in a matrix are represented as colors.

```python
flights=sns.load_dataset("flights")
```

```python
flights.head()
```

```python
sns.catplot(x='month',y='passengers',kind='bar',data=flights)
```

```python
# now as you can see that the data is close to each other so to overcome this problem we will use **split='true'**
sns.barplot(x='month', y='passengers', data=flights, dodge=False)
```

```python
x=flights.pivot_table(index='year',columns='month',values='passengers',aggfunc='sum')
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x,cbar=False)
```

```python
sns.heatmap(x,cbar=False,linewidth=0.5)
```

```python
sns.heatmap(x,cbar=False,linewidth=0.5 , annot=True,fmt='d')
```

```python
sns.heatmap(x,cbar=False,linewidth=0.8 , annot=True,fmt='d',cmap='summer')
#cmap='summer', cmap='winter' it defines the color
# the line width increses the gap between the grids or squares
```

# Use of Annot and fmt

<b> in the above command we saw that we used <u>annot</u> so the Seaborn heatmap, the "annot" parameter allows you to annotate the cells with the numeric value of each cell.
    and Seaborn's heatmap, the <u>fmt parameter</u> is used to format the annotations that appear on the heatmap. It allows you to specify the format string for the numerical values displayed in the cells.

For example, if you want to display the values as percentages, you can use fmt='.2f' to format them with two decimal places.</b>

<b>NOTE</b>:- It is recommended that you should use <u> Annot and fmt</u> together because (you can see yourself what happens when you use just annot and not fmt)

# Clustermap

# Jointplot

<b> Jointplot creates a multi-panel figure that shows both the bivariate (or joint) relationship between two variables along with the univariate (or marginal) distribution of each on seperate axes
    
    We can use jointplots in the followinf 4 manners:-
    1.Scatterplots
    2.Hexbinplots
    3.kdePlots
    4.Regplot

```python
tips=sns.load_dataset('tips')
```

```python
sns.jointplot(x='total_bill',y='tip',kind='hex',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',kind='kde',data=tips)
```

# Distplot

<b> Displots are used to do univariate analysis on a continuous/numerical data.
    
    Distplot combines 3 things in one graph:-
    1. Histogram
    2. Kernel density estimation( PDF)
    3. Rugplot
    

```python
titanic=sns.load_dataset('titanic')
titanic.head()
```

```python
titanic['age'].fillna(titanic['age'].mean(),inplace=True)
```

# Plotting a Distplot

```python
sns.distplot(titanic['age'])
```

# Plotting only on histogram

```python
sns.distplot(titanic['age'],kde=False,bins=100)
#bins is optiona;
```

```python
sns.distplot(titanic['age'],hist=False)
```

# Plotting on rugplot

```python
sns.distplot(titanic['age'],hist=False,kde=False,rug=True)
```

# How to use Distplot

```python
sns.distplot(titanic[titanic['survived']==1]['age'])# blue=survived
sns.distplot(titanic[titanic['survived']==0]['age'])#red =dead
```
